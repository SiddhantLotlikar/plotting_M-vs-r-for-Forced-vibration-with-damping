```python
import numpy as np #arrays and matrices
import matplotlib.pyplot as plt #vizualizations
import math # mathematics library
import seaborn as sns # graphics library
```

```python
zeta_values = [0, 0.1, 0.2, 0.4, 0.8, 1, 2, 4, 8]
```

```python
def amplification_ratio(r,z):
    return 1/math.sqrt((1-r**2)**2 + (2*z*r)**2)

r = np.linspace(0, 2.5, 100) #undamped natural freq range
```

```python
plt.figure(figsize=(10, 6))

for z in zeta_values:
    M = [amplification_ratio(i, z) for i in r]
    plt.plot(r, M, marker='', label=f'zeta={z}')
    plt.xlabel('r')
plt.ylabel('M')
plt.title('Amplification Ratio vs r for Different Zeta Values')
plt.legend()
plt.show()
```
