```python
import numpy as np #arrays and matrices
import matplotlib.pyplot as plt #vizualizations
import math # mathematics library
import seaborn as sns # graphics library
```

```python
z= float(input("enter value of zeta (damping ratio):"))
```

```python
def amplification_ratio(r,g):
    return 1/math.sqrt((1-r**2)**2 + (2*z*r)**2)

r = np.linspace(0, 2.5, 100) #undamped natural freq range


M=[]
for i in r:
    M.append(amplification_ratio(i,z))
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=r, y=M, marker='o')
plt.xlabel('r')
plt.ylabel('M')
plt.title('M vs r')
plt.show()
```
